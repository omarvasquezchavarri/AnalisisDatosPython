```python
#funciones estadisticas con nympy
m1 = np.random.randint(9,size=(3,3))
print(m1)
print()
print(np.amax(m1))
print()
print(np.amin(m1))
print()
print(np.amax(m1,axis=0))
print()
print(np.amax(m1,axis=1))
```

```python
print(np.ptp(m1))
print()
print(np.percentile(m1,50,axis=1))
```

```python
#recorrer matriz de dos dimensiones
for fila in m1:
    for elemento in fila:
        print(elemento, end=" ")


```

```python
print(np.mean(m1,axis=0))
print()
print(np.median(m1,axis=1))
```

```python
m1D = np.array([2,4,3,7,4,8])
print(m1D)
print()
print(np.average(m1D))
print()
print(np.average(m1D,axis=0))
print()
print(np.std(m1D))
```

```python
#Algebra Lineal
m = np.array([[1,-1,2],[3,2,0]])
print(m)
print()
print(m.T)

print()
m1 = np.array([[1],[2],[3]])
print(m1)
print()
print(np.transpose(m1))

```

```python
# operaciones con matrices
m1 = np.array([[2,1,-2],[3,0,1],[1,1,-1]])
print(m1)
print()
m2 = np.array([[-3],[5],[-2]])
print(np.transpose(m2))
print()

x = np.linalg.solve(m1,m2)

print(x)
```
