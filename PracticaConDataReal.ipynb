```python
import pandas as pd
import numpy as np

# 1. Tu diccionario de datos
data2 = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Marta', 'Sofia'],
    'Salario': [30000, 40000, 50000, 35000, 32000],
    'edad': [25, 30, np.nan, 28, 22],
    'ciudad': ['Madrid', 'Barcelona', 'Valencia', None, 'Bilbao']
}

# 2. Crear el DataFrame
df = pd.DataFrame(data2)

# 3. Agregar el índice como una columna llamada 'Id_Indice'
df['Id_Indice'] = df.index

# Mostrar el resultado
print(df)
```

```python
columnas = ['Id_Indice', 'Nombre', 'Salario', 'edad', 'ciudad']
df = df[columnas]
print(df)
```
