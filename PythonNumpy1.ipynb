### filtrado y Visualización de Datos

```python
df2.head(10)
```

```python
#VEntas por Origen Moneda
totalOnline = df2[df2['Origen'].isin(['Online'])]
VO = sum(totalOnline['Total'])
print(f'La suma de ventas Online es {VO}')

totalFisica= df2[df2['Origen'].isin(['Tienda Física'])]
VE = sum(totalFisica['Total'])
print(f'La suma de ventas Tienda Física es {VE}')

x = ['Online','Fisica']
y = [VO,VE]

fig, ax = plt.subplots()

ax.bar(x,y,linewidth=5)
#plt.show()
```

```python
totalOrigen = df2[['Origen','Total']]
totalOrigen2 = totalOrigen.groupby(by=['Origen']).sum()
totalOrigen2

fig, ax = plt.subplots()
valores = totalOrigen2.Total
#tituloM = totalOrigen2.index
tituloM = df2['Origen'].unique()
colores = ['#619cff','#00ba38']

ax.bar(x=tituloM, height=valores,color=colores)
plt.show()

```

```python
df2[['Origen','Cantidad']]
```

```python
CantidadOrigen2 = df2[['Origen','Cantidad']].groupby(by=['Origen']).count()

print(CantidadOrigen2)
fig, ax = plt.subplots()
valores = CantidadOrigen2.Cantidad
#tituloM = totalOrigen2.index
tituloM = df2['Origen'].unique()
colores = ['#619cff','#00ba38']

ax.bar(x=tituloM, height=valores,color=colores)
plt.show()
```

```python
#Cantidad de Unidades que se vendieron por origen
cantidadUniOrigen = df2[['Origen','Cantidad']]
cantidadUniOrigen = cantidadUniOrigen.groupby(by=['Origen']).sum()
print(f'la cantidad de uniades por origen fueron {cantidadUniOrigen}')

fig, ax = plt.subplots()
valores = cantidadUniOrigen.Cantidad
#tituloM = totalOrigen2.index
tituloM = df2['Origen'].unique()
colores = ['#619cff','#00ba38']

ax.bar(x=tituloM, height=valores,color=colores)
plt.show()
```

```python
cantidadUniOrigen = df2[['Origen','Cantidad']]
cantidadUniOrigen = cantidadUniOrigen.groupby(by=['Origen']).sum()
print(f'la cantidad de uniades por origen fueron {cantidadUniOrigen}')

fig, ax = plt.subplots()
valores = cantidadUniOrigen.Cantidad
#tituloM = totalOrigen2.index
tituloM = df2['Origen'].unique()

explode = [0.1,0]
ax.pie(valores,labels=tituloM,explode = explode, autopct='%1.1f%%')
plt.show()
```

```python
#Productos mas vendidos
#df2.head(4)
productosMasVendidos = df2[['Producto','Total']]
productosMasVendidos2 = productosMasVendidos.groupby(by='Producto').sum()
#productosMasVendidos2.sort_values

fig, ax = plt.subplots()

Valores = productosMasVendidos2['Total']
Titulos = productosMasVendidos2.index

# 4. Dibujar las barras y GUARDARLAS en una variable
barras = ax.bar(x=Titulos, height=Valores, color="#0f2ebb")
# 5. 🔥 LA MAGIA: Agregar las etiquetas de datos
ax.bar_label(barras, padding=1) # 'padding=3' separa el número un poquito de la barra
# Opcional: Rotar los nombres si se enciman unos con otros
plt.xticks(rotation=45, ha='right')
# Ajustar el diseño para que no se corte nada y mostrar
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
explode = [0.1,0.1,0.1,0]
ax.pie(Valores,labels=Titulos, explode=explode, autopct='%1.1f%%')
plt.show()
```

```python
# 1. Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 5))

# 2. Dibujar el gráfico de líneas
# - 'marker="o"': añade un punto en cada dato
# - 'linewidth=2': define el grosor de la línea
# - 'color="#619cff"': mantiene la estética azul
ax.plot(Titulos, Valores, marker='o', linewidth=2, color='#619cff')

# 3. Agregar etiquetas de datos sobre cada punto (opcional pero muy útil)
for i, valor in enumerate(Valores):
    ax.annotate(f'{valor}', 
                (Titulos[i], Valores[i]), 
                textcoords="offset points", 
                xytext=(0,10), # Separa el texto 10 puntos arriba del punto
                ha='center')   # Centra el texto

# 4. Personalizar el gráfico
ax.set_title('Tendencia de Ventas por Producto', fontsize=14, fontweight='bold')
ax.set_xlabel('Productos', fontsize=12)
ax.set_ylabel('Cantidad Vendida', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5) # Añade una cuadrícula sutil de fondo

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
