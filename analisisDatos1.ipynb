```python
#Cargamos el dataset
data = {'nombres': ['Ana', 'Luis', 'Carlos', 'Marta', 'Sofia','Miguel'],
        'ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Madrid','Zaragoza'],
        'edades': [25, 30, 35, 28, 22,29],
        'puntuaciones': [85, 90, 95, 88, 92,87]
        }
df5 = pd.DataFrame(data)
df5
```

```python
#agrupar por ciudad
agrupado = df5.groupby('ciudad')
print(agrupado.groups)
```

```python
#calcular suma de edades y puntuaciones por ciudad
dataAgregada = agrupado.agg({'edades': 'mean', 'puntuaciones': 'sum'})
dataAgregada
```

```python
#funciones de agregacion personalizadas
def rangoEdad(serie):
    return serie.max() - serie.min()

#aplicar funcion
dataAgregadaPersonalizada = agrupado.agg({'edades': rangoEdad, 'puntuaciones': rangoEdad})
dataAgregadaPersonalizada
```
