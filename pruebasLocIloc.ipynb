```python
import pandas as pd

df = pd.DataFrame({'Precio_Base': [100, 250, 50]})

# Aplicamos una función lambda a cada fila de la columna
df['IGV'] = df['Precio_Base'].apply(lambda x: x * 0.18)

print(df)
```

```python
productos = [('Celular', 15), ('Laptop', 5), ('Tablet', 30)]

# Le decimos a sorted que use como clave de ordenamiento el elemento en la posición 1
productos_ordenados = sorted(productos, key=lambda prod: prod[1])

print(productos_ordenados)
```
